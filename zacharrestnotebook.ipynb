## Final Project Individual Notebook

For my group's final project, we decided to rank the best neighborhood in Pittsburgh based off of police intervention in the area. This analysis includes things like the arrest records, non-traffic citations, and the police incident blotter. In this notebook, I will be analyzing data that shows the amount of arrests documented throughout Pittsburgh.

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
arrests = pd.read_csv("https://data.wprdc.org/datastore/dump/e03a89dd-134a-4ee8-a2bd-62c40aeebc6f")
num_arrests = arrests.loc[:, "INCIDENTNEIGHBORHOOD"]

arrestdict = dict()

for INCIDENTNEIGHBORHOOD in num_arrests:
    if INCIDENTNEIGHBORHOOD not in arrestdict:
        arrestdict[INCIDENTNEIGHBORHOOD] = 1
    else:
        arrestdict[INCIDENTNEIGHBORHOOD] = arrestdict[INCIDENTNEIGHBORHOOD] + 1

print(arrestdict)

plt.figure(figsize=(20, 3))  # width:20, height:3
plt.bar(range(len(arrestdict)), arrestdict.values(), align='edge', width=0.3)
plt.xticks(range(len(arrestdict)), arrestdict.keys(), rotation='vertical')
plt.show()
```

We can see here that Regent Square, East Carnegie, Central Northside, Herrs Island, Oakwood, St. Clair, Mt. Oliver, Summer Hill, New Homestead, Ridgemont, Swissheim Park, Hays, and Chartiers City are among the lowest arrest rates in Pittsburgh. Of all of these, Mt. Oliver Neighborhood had the lowest number of arrests, making it my nomination for the best neighborhood in Pittsburgh.
